```python
import random

class Person(object):
    def __init__(self):
        self.status = "Susceptible"
        self.days_sick = 0
        self.immune_system = 0
        self.sickness_rate = 10
        
    def infect(self, number_of_days_sick):
        self.status = "Infected"
        self.days_sick = number_of_days_sick
    
    def recover(self):
        self.status = "Recovered"
        self.days_sick = 0
    
    def update(self):
        if(self.status == "Susceptible"):
            self.roll_for_infection()
        elif (self.status == "Infected"):
            self.days_sick -= 1
            if (self.days_sick < 0):
                self.recover()
            
    def roll_for_infection(self):
        is_sick = (random.randint(1,100) <= self.sickness_rate)
        if (is_sick):
            self.infect(5)
            
def
```

```python
my_population = []
currently_infected = []
count = 0
num_of_interaction = 10

for i in range(0,100):      #populates the my_population list
    generic_dude = Person()
    diet = random.randint(1,10)
    sleep = random.randint(1,10)
    exercise = random.randint(1,10)
    age = random.randint(1,10)
    generic_dude.immune_system = diet + sleep + exercise + age
    my_population.append(generic_dude)
    

for z in range(0,10):
    infected_person = random.randint(0,len(my_population)-1)
    
    while(infected_person in currently_infected):
        infected_person = random.randint(0,len(my_population)-1)
        
    my_population[infected_person].infect(5)
    currently_infected.append(infected_person)

for j in range(0, len(my_population)):
    if(my_population[j].immune_system < 20 and j not in currently_infected):
        print("Person", j, "has a poor diet. Chances of getting sick:", 100 - (my_population[j].immune_system/40)*100, "%")


print("Day 1, people sick:", currently_infected)


for person in range(0,len(my_population)):
    interacted = []
    
    if(my_population[person].immune_system < 20 and person not in currently_infected):
        my_population[person].sickness_rate = 60
    else:
        my_population[person].sickness_rate = 10
        
    for j in range(0,num_of_interaction):
        some_person = random.randint(0,len(my_population)-1)
        while(person == some_person):
            some_person = random.randint(0,len(my_population)-1)


        if(my_population[person].status == "Infected"):
            my_population[some_person].roll_for_infection()
            
            if (my_population[some_person].status == "Infected"):
                
                if(some_person not in currently_infected):
                    currently_infected.append(some_person)
                  
        elif (my_population[person].status != "Infected"):
            if(some_person in currently_infected):
                my_population[person].roll_for_infection()

            
        interacted.append(some_person)
            
    print("Person",person, "interacted with persons", interacted)

for person in my_population:
    if(person.status == "Infected"):
        count += 1

print("Total people that got sick:", len(currently_infected))
print(currently_infected)
```
