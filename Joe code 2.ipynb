```python
import random as rd
 
class Person(object):
    def __init__(self):
        self.status = "Susceptible"
        self.days_sick = 0
        self.sickness_rate = 10
        
    def infect(self, number_of_days_sick):
        self.status = "Infected"
        self.days_sick = number_of_days_sick
        self.was_sick = 0
    
    def recover(self):
        self.status = "Recovered"
        self.days_sick = 0
    
    def update(self):
        if(self.days_sick >= 1):
            self.days_sick -= 1
        elif (self.days_sick <= 0):
            self.recover()
 
Joe = Person()
Jane = Person()
day = 0
flag = 0
 
while (Jane.status != "Recovered"):
    
    while(Joe.status != "Infected" and Joe.status != "Recovered"):
        day += 1
        
        if (Joe.status != "Infected" and Joe.status != "Recovered"):
            chance_Joe_getting_sick = rd.randint(1,100) #roll the die for Joe
            
            if(chance_Joe_getting_sick < Joe.sickness_rate): #did Joe get sick
                Joe.infect(5)
        print("Joe is", Joe.status, "day:", day)
        print("Jane is", Jane.status, "day:", day)
 
    day += 1

    if (Joe.status == "Infected" or Joe.status == "Recovered"):
        Joe.update()
        if (Jane.status != "Infected"):
            chance_Jane_getting_sick = rd.randint(1,100)  #roll the die for Jane
 
        if (chance_Jane_getting_sick < Jane.sickness_rate and flag == 0):
            Jane.infect(5)
            flag = 1
            
    if (Jane.status == "Infected" and flag == 1):
        Jane.update()
 
    print("Joe is", Joe.status, "day:", day)
    print("Jane is", Jane.status, "day:", day)
```
